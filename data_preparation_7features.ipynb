```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
total = pd.read_csv('./total_16features.csv')
```

```python
features = ['SPEED', 'MAF', 'ENGINE_RPM',  'THROTTLE_POS', 'INTAKE_MANIFOLD_PRESSURE', 'ENGINE_LOAD', 'FUEL_RATE', 'FUEL_CONSUMPTION']
total = total[features]
total.info()
```

```python
plt.figure(figsize = (25,8))
sns.heatmap(total.iloc[:,:].corr(),annot=True)
```

```python
# split shuffled train and test sets
train_shuffled = total.iloc[:10000, :]
test_shuffled = total.iloc[10000:, :]
```

```python
total.to_csv("{}.csv".format('total_7features'), index = False)
train_shuffled.to_csv("{}.csv".format('train_shuffled_7features'), index = False)
test_shuffled.to_csv("{}.csv".format('test_shuffled_7features'), index = False)
```
